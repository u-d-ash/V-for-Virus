```python
import pandas as pd
```

```python
data = pd.read_csv("data/covid_19_data.csv")
```

```python
data
```

```python
india_data = data[data['Country/Region'] == "India"]
```

```python
india_data
```

```python
india_data = india_data.drop(['SNo', 'Province/State', 'Country/Region', 'Last Update'], axis = 1)
```

```python
india_data['Date'] = pd.to_datetime(india_data['ObservationDate'])
```

```python
india_data = india_data.groupby('Date', as_index=False).sum()
```

```python
india_data
```

```python
train_data = india_data[: 375]
test_data = india_data[375 : ]
```

```python
train_data.to_csv("data/training.csv", index=False)
test_data.to_csv("data/testing.csv", index = False)
```
